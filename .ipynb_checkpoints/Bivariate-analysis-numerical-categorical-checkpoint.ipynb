```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head()
```

```python
sns.barplot(data=df,x="neighbourhood_group",y="price")
```

```python
sns.barplot(data=df,x="neighbourhood_group",y="number_of_reviews",estimator="sum")
```

```python
sns.barplot(data=df,x="neighbourhood_group",y="price",hue="room_type")
```

```python
sns.boxplot(data=df,x="neighbourhood_group",y="number_of_reviews")
```

```python
sns.violinplot(data=df,x="neighbourhood_group",y="price")
```

```python
sns.distplot(df[df["neighbourhood_group"]=="Queens"]["price"])
```
